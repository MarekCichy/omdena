```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_rows', 1000)
pd.set_option('max_colwidth', -1)
```

```python
df = pd.read_excel('zapytaj_1_16_20.xlsx')
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0', 'Unnamed: 8'], inplace=True)
```

```python
df.head()
```

```python
df["description"] = df["description"].replace(regex={'\n': ''})
```

```python
df.tail()
```

```python
splitlabels = df.label.astype('str').apply(lambda x:x.split(', '))
```

```python
labellist = splitlabels.tolist()
```

```python
result = set(x for l in labellist for x in l)
```

```python
result
```

```python
for i in ['','none','nan','advice']:
    result.discard(i)
```

```python
label_list=list(result)
label_list.sort()
label_list
```

```python
df.label = df.label.fillna(0)
for i in label_list:
    df[i]=df.label.str.contains(i)

for i in label_list:
    df[i]=df[i].fillna(0)
    df[i]=df[i].astype(int)
```

```python
stats = df.loc[:,'aesthetic_preference':].sum(axis=0)
```

```python
stats.plot(kind='barh')
```

```python
rowsums = df.loc[:,'aesthetic_preference':].sum(axis=1)
x=rowsums.value_counts()


ax = sns.barplot(x.index, x.values)
plt.title("Multiple categories per annotation")
plt.ylabel('# of annotations', fontsize=12)
plt.xlabel('# of categories', fontsize=12)
```

```python
df.loc[:,['title','description','label']][rowsums>1]
```

```python
#df_w_dates =  df[pd.notna(df.date_posted)][['date_posted']+categories]
select_cols = label_list
df_w_dates =  df[pd.notna(df.date_posted)][select_cols+['date_posted']]
df_w_dates['datetime'] = pd.to_datetime(df_w_dates.date_posted)
df_w_dates['month'] = df_w_dates.datetime.apply(lambda x: x.month) 
df_w_dates['year'] = df_w_dates.datetime.apply(lambda x: x.year)
df_w_dates['month_year'] = df_w_dates.datetime.dt.to_period('M')
df_dates_grouped = df_w_dates.drop(columns=['month','year']).groupby(['month_year']).sum()
```

```python
df_dates_grouped.columns
```

```python

melted_date_df = df_dates_grouped.reset_index().melt(id_vars = 'month_year', value_vars= select_cols)

melted_date_df[['month_year','value','variable']].set_index('month_year', inplace=True)

#fix error by sorting by the time period, converting to strings and then plotting with seaborn
#sns.tsplot(time = 'month_year', value='value', hue='variable', data= melted_date_df.reset_index())
```

```python
plt.figure(figsize=(15,5))
melted_date_df.sort_values(by='month_year', inplace=True)
melted_date_df.month_year = melted_date_df.month_year.astype('str')
chart = sns.lineplot(x = 'month_year', y='value', hue='variable', data= melted_date_df)
#chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)
plt.show()
#This is obviously fairly arbitrary because of the data selections we all worked on
```
